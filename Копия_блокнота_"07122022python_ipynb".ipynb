```python
df1=pd.DataFrame({'sex':np.random.choice(['male','female'],p=(0.3,0.7),size=1000)})
df1['age']=np.where(df1.sex=='male',np.round(np.random.normal(35,3),0),np.round(np.random.normal(25,3),0))
df1['income']=np.where((df1.sex=='male')&(df1.age>37),np.round(np.random.normal(100000,20000),0),np.round(np.random.normal(50000,10000),0))
df1['social']=np.random.choice(['married','divorced'],size=1000)
```

```python
df1[(df1.sex=='male')&(df1.social=='married')]['income']
```

```python
fig,ax=plt.subplots(2,2,figsize=(16,4))
df1.loc[(df1.sex=='male')&(df1.social=='married'),'income'].hist(ax=ax[0,0])
ax[0,0].set_title('male-married')
df1.loc[(df1.sex=='male')&(df1.social=='divorced'),'income'].hist(ax=ax[0,1])
df1.loc[(df1.sex=='female')&(df1.social=='married'),'income'].hist(ax=ax[1,0])
df1.loc[(df1.sex=='female')&(df1.social=='divorced'),'income'].hist(ax=ax[1,1])
```

```python
sns.heatmap(pd.pivot_table(df1.groupby(['sex','social'])['income'].\
               mean().reset_index(),
               index=['sex'],
               columns=['social'],
               values=['income']))
```

```python
sns.catplot(data=df1,x='income',y='income',hue='sex')
```

```python
fig,ax=plt.subplots(df1.sex.nunique(),df1.social.nunique(),figsize=(8,8))
for i,sex in enumerate(df1.sex.unique()):
  for j,social in enumerate(df1.social.unique()):
    df1[(df1.social==social)&(df1.sex==sex)]['age'].hist(ax=ax[i,j])
    ax[i,j].set_title(social+' '+sex)
```
